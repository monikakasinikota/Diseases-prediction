```python
import numpy as np
import pandas as pd
from imblearn.under_sampling import TomekLinks
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
diabetes_dataset=pd.read_csv("../datasets/diabetes.csv")
```

```python
x=diabetes_dataset.drop(columns=['Outcome'],axis=1)
y=diabetes_dataset['Outcome']
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Initialize Tomek Links
tomek = TomekLinks()

# Apply Tomek Links undersampling to remove unnecessary majority class points
X_resampled, y_resampled = tomek.fit_resample(x_train, y_train)

# Print class distribution before and after undersampling
print("Original class distribution:\n", y_train.value_counts())
print("\nClass distribution after Tomek Links:\n", pd.Series(y_resampled).value_counts())
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.head()
```

```python
# Train a model (Random Forest in this example
X_resampled = np.array(X_resampled)
X_test = np.array(x_test)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_resampled, y_resampled)

# Make predictions
y_pred = model.predict(x_test)
```

```python
# Compute accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Classification report (Precision, Recall, F1-score)
print("Classification Report:\n", classification_report(y_test, y_pred))
```
